```python
import json
```

```python
f = open('../data/h.all.v2023.1.Hs.json')

data = json.load(f)
```

```python
for d in data:
    print(d)
    print(data[d]['geneSymbols'])
    print(data[d])
#    print(data[d]['geneSymbols'])
```

```python
for d in data:
    with open(d+'.yaml', 'w') as f:
        f.write('name: '+d+'\n')
        f.write('gene_symbols: []'+'\n')
        f.write('gene_ids:'+'\n')
        for g in data[d]['geneSymbols']:
            f.write(' - '+g+'\n')   
        f.write('taxon: human'+'\n')
        f.write('description: null')
        f.close()
```
